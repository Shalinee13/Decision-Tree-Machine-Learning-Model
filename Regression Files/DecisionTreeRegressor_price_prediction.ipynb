```python
X=data[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
```

```python
Y=data["price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

## DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model=DecisionTreeRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
y_predict
```

```python
model.score(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

## Accuracy

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_test, y_predict)*100
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mean_squared_error(Y_test, y_predict)
```

```python
np.sqrt(mean_squared_error(Y_test, y_predict))
```

```python
mean_absolute_error(Y_test, y_predict)
```

```python
path=model.cost_complexity_pruning_path(X_train,Y_train)
```

```python
path
```

```python
alphas=path['ccp_alphas']
```

```python
alphas
```

```python
len(alphas)
```

```python
alphas.min()
```

```python
alphas.max()
```

```python
accuracy_train=[]
accuracy_test=[]

for i in alphas:
    model=DecisionTreeRegressor(ccp_alpha=i)
    model.fit(X_train,Y_train)
    y_predict_train=model.predict(X_train)
    y_predict_test=model.predict(X_test)
    
    accuracy_train.append(r2_score(Y_train, y_predict_train))
    accuracy_test.append(r2_score(Y_test, y_predict_test))
    
```

```python
accuracy_train
```

```python
print(accuracy_test)
```

```python
sns.lineplot(y=accuracy_train,x=alphas)
sns.lineplot(y=accuracy_test,x=alphas)
plt.show()
```
